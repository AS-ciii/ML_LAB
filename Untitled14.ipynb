```python
import pandas as pd
data = pd.read_csv()
print(data)

def find_s_algorithm(data):
  attributes = data.iloc[:, :-1]
  target = data.iloc[:, -1]

  hypothesis = ['$']*len(attributes[0])

  for i in range(len(target)):
    if target[i] == "Yes":
      hypothesis = attributes[i].copy()
      break

  for i in range(len(target)):
    if target[i] == "Yes":
      for j in range(len(hypothesis)):
        if hypothesis[j] != attributes[i][j]:
          hypothesis[j] = '?'

  return hypothesis

final_hypo = find_s_algorithm(data)
print("Most specific hypothesis:", final_hypo)
```

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

np.random.seed(42)

values = np.random.rand(100)
labels = []

for i in values[:50]:
  if i <=0.5:
    labels.append("class1")
  else:
    labels.append("class2")

labels += [np.nan]*50

print(labels)

Data = {
    "Point" : [f"x{i+1}" for i in range(100)],
    "Value" : values,
    "Label" : labels
}

df = pd.DataFrame(Data)
df.head()

labeled_df = df[df["Label"].notna()]
x_train = labeled_df[["Value"]]
y_train = labeled_df["Label"]

unlabeled_df = df[df["Label"].isna()]
x_test = unlabeled_df[["Value"]]

k_vals = {1, 2, 3, 4, 5, 20, 30}
results = {}

for k in k_vals:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train, y_train)
  predictions = knn.predict(x_test)
  results[k] = predictions
  unlabeled_df[f"Label_k{k}"] = predictions
print(predictions)

df1 = unlabeled_df.drop(columns=["Label"], axis=1)
print(df1)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt

data = load_breast_cancer()
x = data.data
y = data.target

model = KMeans(n_clusters=2)
model.fit(x, y)
labels = model.labels_

pca = PCA(n_components=2)
x_Reduced = pca.fit_transform(x)

plt.figure(figsize=(8, 6))
plt.title("K Means")
plt.xlabel("pca 1")
plt.ylabel("pca 2")
plt.scatter(x_Reduced[:, 0], x_Reduced[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.show()

adj_labels = 1-labels

acc1 = accuracy_score(y, labels)
acc2 = accuracy_score(y, adj_labels)

print("Max clustering acc:", max(acc1, acc2))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt

iris = load_iris()
x = iris.data
y = iris.target

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)

plt.figure(figsize=(8,6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y, cmap='viridis', edgecolor = 'k', alpha=0.6)
plt.title("pca")
plt.xlabel("pca 1")
plt.ylabel("pca 2")
plt.colorbar(label="target classes")
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier, plot_tree

import matplotlib.pyplot as plt

data = load_breast_cancer()
x = data.data
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

plt.figure(figsize=(10, 6))
plot_tree(
    model,
    feature_names = data.feature_names,
    class_names = data.target_names,
    rounded = True,
    filled = True,
    fontsize = 10
)
plt.title("Decision Tree")
plt.show()

```

```python
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_olivetti_faces

faces = fetch_olivetti_faces()
x = faces.data
y = faces.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)

nb = GaussianNB()
nb.fit(x_train, y_train)

y_pred = nb.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

for i in range(5):
  plt.imshow(x_test[i].reshape(64, 64), cmap='gray')
  plt.title(f"Actual : {y_test[i]}, Predicted : {y_pred[i]}")
  plt.axis('off')
  plt.show()
```

```python
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
x = housing.data[:, housing.feature_names.index("AveRooms")].reshape(-1, 1)
y = housing.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

plt.title("Linear Regression")
plt.xlabel("AveRooms")
plt.ylabel("House Price")
plt.scatter(x_test, y_test, color="red", label="actual")
plt.plot(x_test, y_pred, color="blue", label="pred")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
data = pd.read_csv(url)

data = data.dropna()

x = data[["horsepower"]].values
y = data["mpg"].values


poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x_poly, y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(x_train, y_train)

x_test_orig = x_test[:, 1]
y_pred = model.predict(x_test)


plt.title("Polynomial Regression")
plt.xlabel("horsepower")
plt.ylabel("mgp")
plt.scatter(x_test_orig, y_test, color="green", label="actual")
plt.scatter(x_test_orig, y_pred, color="purple", label="pred")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore")

from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)

df.head()
df.info()
df.nunique()

df.isnull().sum()


df.describe().T

Numerical = df.select_dtypes(include=[np.number]).columns
print(Numerical)

for col in Numerical:
  plt.figure(figsize=(10, 6))
  df[col].plot(kind='hist', title=col, edgecolor='black')
  plt.ylabel("freq")
  plt.show()

for col in Numerical:
  plt.figure(figsize=(10, 6))
  sns.boxplot(df[col], color="blue")
  plt.ylabel(col)
  plt.title(col)
  plt.show()

outliters_summary = {}

for feature in Numerical:
  q1 = df[feature].quantile(0.25)
  q3 = df[feature].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = (df[feature]<lower_bound) | (df[feature]>upper_bound)
  outliters_summary[feature] = len(outliers)
  print(f"{feature} : {outliers.sum()}")
```
